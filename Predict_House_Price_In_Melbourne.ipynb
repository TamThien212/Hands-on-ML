```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])
#delete column axis = 1
num_attribs = list( data_copy.drop(["Suburb","Type","Address","Method","SellerG","Date","CouncilArea","Regionname","Price"],axis = 1))
cat_attribs = list(cat)

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(data,test_size=0.2,random_state=42)
```

```python
data_train = train_set.drop("Price",axis = 1)
label_train = train_set["Price"]
data_test = test_set.drop("Price",axis = 1)
label_test = test_set["Price"]
```

```python
data_train_prepare = full_pipeline.fit_transform(data_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(data_train_prepare,label_train)
```

```python
# let's try the full preprocessing pipeline on a few training instances
some_data = data_train.iloc[:5]
some_labels = label_train.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print("Predictions:", lin_reg.predict(some_data_prepared))
```

```python
print("Labels:", list(some_labels))
```

```python
lin_reg.score(data_train_prepare,label_train)
```

```python
from sklearn.metrics import mean_squared_error

predictions = lin_reg.predict(data_train_prepare)
lin_mse = mean_squared_error(label_train, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
lin_scores = cross_val_score(lin_reg, data_train_prepare, label_train,
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
data_test_prepare = full_pipeline.transform(data_test)
predictions_test = lin_reg.predict(data_test_prepare)
lin_mse = mean_squared_error(label_test, predictions_test)
lin_rmse_test = np.sqrt(lin_mse)
lin_rmse_test
```

```python
lin_reg.score(data_test_prepare,label_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(data_train_prepare,label_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, data_train_prepare, label_train,
scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
display_scores(rmse_scores)
```
